```python
from googletrans import Translator
from gtts import gTTS
import IPython.display as ipd
import speech_recognition as sr
import os
import pyttsx3


tr = Translator()
```

```python
x="Hello. My name is Pearl."
l='en'
```

```python
lan=[]
for i in range (3):
    lan.append(input("Select Language:"))
```

```python
r = sr.Recognizer()
with sr.Microphone() as source:
    print("Say something!")
    audio = r.listen(source)    
    try: 
        x = r.recognize_google(audio) 
        print("you said: "+ x)  
      
    except sr.UnknownValueError: 
        print("Google Speech Recognition could not understand audio") 
    except sr.RequestError as e: 
        print("Could not request results from GoogleSpeech Recognition service; {0}".format(e))
              
l=tr.detect(x)
l=l.lang
```

```python
y=[]
for i in range(3):   
    y.append(tr.translate(x, src=l, dest=lan[i]))
    print(y[i].text)

for i in range(3):    
    tts = gTTS(y[i].text, lang=lan[i])
    name=lan[i]+'.mp3'
    tts.save(name)
```
